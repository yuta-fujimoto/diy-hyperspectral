```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
```

```python
image_ref = plt.imread('./0.png')
print(image_ref.shape)

ystart, yend = 50, 187
xstart, xend = 97, 305

plt.imshow(image_ref[ystart:yend, xstart:xend])
```

```python
imgFiles = glob.glob('[0-9].png')
imgFiles.sort()

for i, f in enumerate(imgFiles):
    image = plt.imread(f)[ystart:yend, xstart:xend]
    plt.imsave(f'{f[:-4]}_crop.png', image)
    if i == 0:
        plt.imshow(image)
        plt.axis('off')
        plt.show()
    
```

```python
wb_ystart, wb_yend = 10, 225
wb_xstart, wb_xend = 40, 360



wb_image = plt.imread('./white_balance.png')
wb_image = cv2.resize(wb_image[wb_ystart:wb_yend, wb_xstart:wb_xend], (xend - xstart, yend - ystart))
plt.imshow(wb_image)
plt.axis('off')
plt.show()

plt.imsave('./white_balance_crop.png', wb_image)
```
